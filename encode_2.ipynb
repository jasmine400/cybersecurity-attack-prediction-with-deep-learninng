```python
# i is a dictionary contains the pulse info with indicator which has been reported by pulse id
# gk is a groupbied df first dataframe which has indicators info
# k is number of indicator in string format
# out put of function is a dictionary i input dictionary which has been updated by a new key role
from numpy import NaN
def get_role(i,k,gk):
  #cr is the time of pulse which has been created for specefic indicator
  cr=i['created']
  #gk_info is a part of groupbied dataframe with k indicator
  gk_info=gk.get_group(k)
  # eq checks the time between groupbied dataframe and the pulse which has been reported indicator
  eq=gk_info.loc[gk_info['created'] == cr]
  # if match happens it retrieves role of the indicator from first groupbied dataframe in the time which has been specified
  if len(eq)>0:
    role=eq.iloc[0]['role']
    dic1={'role':role}
    i.update(dic1)
  else:
    dic1={'role':NaN}
    i.update(dic1)
  # print(i)
  return (i)
```

```python
def times(sampl):
  list_of_times=[]
  length=len(sampl)
  last_day_idx=length-2
  last_day=sampl.iat[last_day_idx,2]
  if last_day:
    for i in range(30):
      # print(last_day)
      # print(i)
      dy=last_day-pd.to_timedelta(i, unit='d')
      # sampl['date1'] = sampl['created'] -  pd.to_timedelta(sampl['diff'], unit='d')
      list_of_times.append(dy)
  return(list_of_times)
```

```python
def add_benign_days(sampl):

  list_of_times = times(sampl)
  # print(list_of_times)
  indicator=sampl.iat[0,1]
  for i in range(len(list_of_times)):
    df_s=sampl.loc[sampl['created'] == list_of_times[i]]
    if len(df_s)<1:
      df2 = { 'id':'nan','indicator': indicator, 'created': list_of_times[i] , 'targeted_countries': [0], 'industries': [0],'pulse':'nan', 'count' : 0, 'attack_type': [0]}
      sampl = sampl.append(df2, ignore_index = True)
  return sampl
```

```python
#pre encode
def pre_encode(atk_type):

  attack_type=[]

  for i in atk_type:
    if 0 in i:
      attack_type.append('0')
    elif 'HoneyNet' in i:
      attack_type.append('HoneyNet')
    elif 'scanning_host' in i:
      attack_type.append('scanning_host')
    elif 'bruteforce' in i:
      attack_type.append('bruteforce')
    elif 'scanning' in i:
      attack_type.append('scanning_host')
    else:
      attack_type.append('other')
  print(len(attack_type))
  return (attack_type)

```

```python
def pre_encode_industries(industry):

  industries=[]
  for i in industry:
    if 0 in i:
      industries.append('0')
    elif len(i)>0:
      print(i)
      industries.append(i[0])
    else :
      industries.append('other')
  return (industries)
```
